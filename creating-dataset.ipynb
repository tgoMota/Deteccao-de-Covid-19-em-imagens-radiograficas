## **Image Cropping**

```python
def cutting_image(image, image_height, image_width, window_size, window_cut):
    windows = []
    for y in range(0, image_height, window_cut):
        for x in range(0, image_width, window_cut):
            window = image[y:y+window_size, x:x+window_size]
            windows.append(window)
    return windows
```

# **Data Augmentation**



```python
#cria cópias com rotação de 90 graus
rot_x = []
rot_y = []
for i in range(len(data_x)):
    rot_x.append(cv2.rotate(data_x[i], cv2.ROTATE_90_COUNTERCLOCKWISE))
    rot_y.append(data_y[i])
```

```python
#cria cópias espelhadas horizontalmente
flip1_x = []
flip1_y = []
for i in range(len(data_x)):
    flip1_x.append(cv2.flip(data_x[i], 1))
    flip1_y.append(data_y[i])
```

```python
#cria cópias espelhadas verticalmente
flip2_x = []
flip2_y = []
for i in range(len(data_x)):
    flip2_x.append(cv2.flip(data_x[i], 0))
    flip2_y.append(data_y[i])
```

```python
#concatena as imagens formadas no array principal
data_x = [*data_x,*rot_x,*flip1_x, *flip2_x]
data_y = [*data_y,*rot_y,*flip1_y, *flip2_y]
```
